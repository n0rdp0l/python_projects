```python
# import necessary modules
from os import link
from bitarray import test
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# retrieve the html content from the given imdb url
html_text = requests.get('https://www.imdb.com/search/title/?title_type=feature&genres=action&explore=genres&view=advanced').text

# parse the html content using BeautifulSoup
soup = BeautifulSoup(html_text, 'lxml')

genre_table = soup.find_all('div', class_ = 'aux-content-widget-2')
table_numbers = genre_table[0].find_all('td')
table_genres = genre_table[0].find_all('a',href = True)
genres = []
numbers = []
for entry, number in zip(table_genres, table_numbers):
    genre = entry.text.lower()
    genres.append(genre)
    number_text = number.text.strip()
    number_digits = ''.join(filter(str.isdigit, number_text))
    number_int = int(number_digits)
    numbers.append(number_int)
```

```python
def loop_through_number(number):
    step = 50
    i = 1
    pages = []
    while (i-1)*step + 1 <= number:
        page = (i-1)*step + 1
        pages.append(page)
        i += 1
    return pages


pages_list = []
for number in numbers:
    pages_list.append(loop_through_number(number))
```
